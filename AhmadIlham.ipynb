```python
#import data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/drive/MyDrive/AhmadIlham TM9 dataset /Iris.csv')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
print(data['Species'].unique())
```

```python

from sklearn.preprocessing import LabelEncoder
for col in ['Id', 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'Species']:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    data
    print (data['Id'],data['SepalLengthCm'],data['SepalWidthCm'],data['PetalLengthCm'],data['Species'])
```

```python
data
```

```python
# Menentukan data x dan y
x = data.drop('Species', axis = 1)
y = data['Species']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state =0)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors =1)
```

```python
knn.fit(x_train, y_train)
```

```python
print("Akurasi score sebesar: {}".format(knn.score(x_test,y_test)*100))
```

```python
y_prediksi = knn.predict(x_test)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_prediksi)
```

```python
print("Prediksi:",y_prediksi)
print("Akurasi: {}%".format(accuracy * 100))
```
